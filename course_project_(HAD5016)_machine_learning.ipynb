**Restricting the sample to only include those with clinical information prior to or within 3 months of Depression diagnosis**

```python
measurement_to_original_date_col = {
    'sBP': 'sBP_Date',
    'BMI': 'BMI_Date',
    'LDL': 'LDL_Date',
    'HDL': 'HDL_Date',
    'A1c': 'A1c_Date',
    'TG': 'TG_Date',
    'FBS': 'FBS_Date',
    'Total_Cholesterol': 'Total_Cholesterol_Date'
}

print("Dictionary 'measurement_to_original_date_col' created:")
print(measurement_to_original_date_col)
```

```python
import pandas as pd

# Correct list of columns to be pivoted (measurements and their dates)
measurement_and_date_cols_to_pivot = [
    'Age_at_Exam', 'sBP', 'sBP_Date', 'BMI', 'BMI_Date', 'LDL', 'LDL_Date', 'HDL', 'HDL_Date',
    'A1c', 'A1c_Date', 'TG', 'TG_Date', 'FBS', 'FBS_Date', 'Total_Cholesterol', 'Total_Cholesterol_Date'
]

# Identify all other columns in diabetes_data that should be treated as static (id_vars for merge)
# Exclude 'Patient_ID' (the index for pivoting) and 'visit_number'
all_cols_in_diabetes_data = diabetes_data.columns.tolist()
id_vars_for_merge = [
    col for col in all_cols_in_diabetes_data
    if col not in measurement_and_date_cols_to_pivot and col not in ['Patient_ID', 'visit_number']
]

# Ensure no duplicates in id_vars_for_merge by converting to set and back to list
id_vars_for_merge = list(set(id_vars_for_merge))

# Pivot the table to wide format, including both measurements and their dates
diabetes_data_wide = diabetes_data.pivot_table(
    index='Patient_ID',
    columns='visit_number',
    values=measurement_and_date_cols_to_pivot
)

# Flatten the MultiIndex columns
diabetes_data_wide.columns = [f'{col[0]}_visit_{col[1]}' for col in diabetes_data_wide.columns]

# Extract unique patient info for static columns using the corrected id_vars_for_merge
# Only include Patient_ID in the selection if it's not already in id_vars_for_merge
# and then add it for the merge key
unique_patient_info_cols = ['Patient_ID'] + id_vars_for_merge
unique_patient_info = diabetes_data.drop_duplicates(subset=['Patient_ID'])[unique_patient_info_cols]

# Merge static info back
diabetes_data_wide = pd.merge(unique_patient_info, diabetes_data_wide, on='Patient_ID', how='left')

print("Head of the diabetes_data_wide DataFrame:")
display(diabetes_data_wide.head())
print("\nShape of the diabetes_data_wide DataFrame:")
print(diabetes_data_wide.shape)
```

```python
import pandas as pd

all_measurement_conditions = []

# Define the 3-month window
THREE_MONTHS = pd.to_timedelta('90 days')

print("Starting filtering process based on measurement dates relative to DM_OnsetDate (including 3 months after diagnosis)...")

# Iterate through each measurement and its corresponding date column
for measurement, original_date_col_name in measurement_to_original_date_col.items():
    # Find all visit numbers available for this measurement
    visit_numbers_for_measurement = sorted(list(set(
        int(col.split('_')[-1])
        for col in diabetes_data_wide.columns
        if col.startswith(f'{measurement}_visit_')
    )))

    # Initialize a boolean Series for the current measurement, initially all False.
    # This will be True for a patient if at least one of their visits meets the criteria for this measurement.
    current_measurement_condition = pd.Series(False, index=diabetes_data_wide.index)

    if not visit_numbers_for_measurement:
        print(f"Warning: No visit-specific columns found for measurement '{measurement}'. Patients will be filtered out if this measurement is required.")
        # If no data exists for a required measurement, no patient can satisfy the 'at least one valid measurement' criteria for it.
        # So, we append an all-False condition to ensure these patients are filtered out in the final 'AND' step.
        all_measurement_conditions.append(pd.Series(False, index=diabetes_data_wide.index, name=f'condition_{measurement}'))
        continue

    # Iterate through all available visits for the current measurement
    for visit_num in visit_numbers_for_measurement:
        m_col = f'{measurement}_visit_{visit_num}'
        d_col = f'{original_date_col_name}_visit_{visit_num}'

        # Check if both the measurement value column and its date column exist for this visit
        if m_col in diabetes_data_wide.columns and d_col in diabetes_data_wide.columns:
            # Condition: date is on or before DM_OnsetDate + 3 months AND measurement value is not NaN AND measurement date is not NaN
            condition_for_this_visit = (diabetes_data_wide[d_col].notna()) & \
                                       (diabetes_data_wide[m_col].notna()) & \
                                       (diabetes_data_wide[d_col] <= diabetes_data_wide['DM_OnsetDate'] + THREE_MONTHS)
            # Update the overall condition for the current measurement type (OR logic for different visits)
            current_measurement_condition = current_measurement_condition | condition_for_this_visit

    all_measurement_conditions.append(current_measurement_condition.rename(f'condition_{measurement}'))
    print(f"Condition for '{measurement}' created. Patients meeting condition: {current_measurement_condition.sum()}")
```

```python
print("\nCombining all individual measurement conditions...")
# Combine all individual measurement conditions using a logical 'OR'
# A patient must satisfy the date condition for *at least one* specified measurement
final_filter_mask = pd.Series(False, index=diabetes_data_wide.index) # Initialize with False for OR operation
for condition_series in all_measurement_conditions:
    final_filter_mask = final_filter_mask | condition_series

# Apply the final filter to the DataFrame
initial_patient_count = diabetes_data_wide.shape[0]
data_copy = diabetes_data_wide.copy()

# Before filtering, ensure 'DM_OnsetDate' is still datetime type
if not pd.api.types.is_datetime64_any_dtype(data_copy['DM_OnsetDate']):
    data_copy['DM_OnsetDate'] = pd.to_datetime(data_copy['DM_OnsetDate'], errors='coerce')

# Re-apply the initial depression filter to data_copy before final_filter_mask
# This step is crucial because data_copy might have lost this filter after reconstruction
condition_no_depression = data_copy['Depression'] == 0
condition_depression_after_diabetes = (data_copy['Depression'] == 1) & (data_copy['DM_OnsetDate'] <= data_copy['Depression_OnsetDate'])

data_copy = data_copy[condition_no_depression | condition_depression_after_diabetes]

# Now apply the measurement date filter
# Ensure final_filter_mask is aligned with data_copy's index before applying
data_copy_filtered = data_copy[final_filter_mask.loc[data_copy.index]].copy()

print("\nFiltering complete.")
print(f"Initial number of patients (before measurement date filter, after depression filter): {data_copy.shape[0]}")
print(f"Number of patients retained after date-based filtering: {data_copy_filtered.shape[0]}")
print(f"Percentage of patients retained: { (data_copy_filtered.shape[0] / data_copy.shape[0]) * 100:.2f}%")
print("\nNew shape of the diabetes_data_wide DataFrame:")
print(data_copy_filtered.shape)
print("\nHead of the filtered diabetes_data_wide DataFrame:")
display(data_copy_filtered.head())
```

**Filtering to keeping only the clinical values for the visit before or the most recent visit after**

```python
import pandas as pd
import numpy as np
import re

# Ensure data_copy_filtered is available and DM_OnsetDate is datetime
# This implicitly relies on prior cells being executed, which the kernel state confirms.
if 'data_copy_filtered' not in locals() or data_copy_filtered.empty:
    raise NameError("data_copy_filtered DataFrame is not available. Please ensure previous steps are executed.")

if not pd.api.types.is_datetime64_any_dtype(data_copy_filtered['DM_OnsetDate']):
    data_copy_filtered['DM_OnsetDate'] = pd.to_datetime(data_copy_filtered['DM_OnsetDate'], errors='coerce')


# Define clinical measurements and their corresponding date columns prefixes
clinical_measurements = {
    'sBP': 'sBP_Date',
    'BMI': 'BMI_Date', # BMI is still a measurement we're processing
    'LDL': 'LDL_Date',
    'HDL': 'HDL_Date',
    'A1c': 'A1c_Date',
    'TG': 'TG_Date',
    'FBS': 'FBS_Date',
    'Total_Cholesterol': 'Total_Cholesterol_Date'
}

# Find the maximum visit number present in the columns to limit iteration
max_visit_num = 0
for col in data_copy_filtered.columns:
    match = re.search(r'_visit_(\d+)$', col)
    if match:
        visit_num = int(match.group(1))
        if visit_num > max_visit_num:
            max_visit_num = visit_num


# Identify static columns (all columns not related to visit-specific measurements or dates)
# This part needs adjustment. 'Age_at_Exam' is a visit-specific column but not part of clinical_measurements for independent best-visit logic.
all_cols = data_copy_filtered.columns.tolist()

# Generate a set of all columns that are *definitely* visit-specific measurements or their dates
visit_value_and_date_cols = set()
for val_col_prefix, date_col_prefix in clinical_measurements.items():
    for i in range(1, max_visit_num + 1):
        visit_value_and_date_cols.add(f'{val_col_prefix}_visit_{i}')
        visit_value_and_date_cols.add(f'{date_col_prefix}_visit_{i}')
# Also explicitly add Age_at_Exam visit columns to this set, as they are not static
for i in range(1, max_visit_num + 1):
    visit_value_and_date_cols.add(f'Age_at_Exam_visit_{i}')


# Now, static_cols are those that are not 'Patient_ID', 'DM_OnsetDate', and are not any of the visit-specific measurement/date columns
static_cols = [
    col for col in all_cols
    if col not in visit_value_and_date_cols and col not in ['Patient_ID', 'DM_OnsetDate']
]


# List to store processed data for each patient
best_visit_rows = []


# Iterate over each patient in the filtered DataFrame
for sequential_idx, (idx_original, patient_data) in enumerate(data_copy_filtered.iterrows()): # Use sequential_idx for clearer progress reporting
    patient_id = patient_data['Patient_ID']
    dm_onset_date = patient_data['DM_OnsetDate']

    # Initialize a dictionary for this patient's best values, starting with static info
    current_patient_best_values = {'Patient_ID': patient_id, 'DM_OnsetDate': dm_onset_date}
    for col in static_cols:
        current_patient_best_values[col] = patient_data[col] # Copy static values

    # This dictionary will store the best visit number for each measurement
    best_visit_num_for_measurement = {}

    # Process each clinical measurement
    for measurement_val_col_prefix, measurement_date_col_prefix in clinical_measurements.items():
        temp_df_data = []
        for visit_num in range(1, max_visit_num + 1):
            val_col_name = f'{measurement_val_col_prefix}_visit_{visit_num}'
            date_col_name = f'{measurement_date_col_prefix}_visit_{visit_num}'

            # Use .get() with None as default to handle missing columns gracefully
            val = patient_data.get(val_col_name)
            date = patient_data.get(date_col_name)

            if pd.notna(val) and pd.notna(date):
                temp_df_data.append({'value': val, 'date': pd.to_datetime(date), 'visit_num': visit_num})

        temp_df = pd.DataFrame(temp_df_data)

        best_value = np.nan
        best_date = pd.NaT
        selected_visit_num = np.nan # Store the visit_num for this measurement

        if not temp_df.empty:
            # Calculate time difference from DM_OnsetDate
            temp_df['time_diff'] = temp_df['date'] - dm_onset_date

            # Filter for visits on or before DM_OnsetDate
            before_dm = temp_df[temp_df['time_diff'] <= pd.Timedelta(0)].copy() # Use .copy() to avoid SettingWithCopyWarning

            if not before_dm.empty:
                # Find the closest visit *before or on* DM_OnsetDate (max date means smallest negative time_diff or zero)
                closest_before_idx = before_dm['time_diff'].idxmax()
                best_value = before_dm.loc[closest_before_idx, 'value']
                best_date = before_dm.loc[closest_before_idx, 'date']
                selected_visit_num = before_dm.loc[closest_before_idx, 'visit_num']
            else:
                # If no visits before or on DM_OnsetDate, find the closest visit *after* DM_OnsetDate
                after_dm = temp_df[temp_df['time_diff'] > pd.Timedelta(0)].copy() # Use .copy()
                if not after_dm.empty:
                    # Find the closest visit *after* DM_OnsetDate (min positive time_diff)
                    closest_after_idx = after_dm['time_diff'].idxmin()
                    best_value = after_dm.loc[closest_after_idx, 'value']
                    best_date = after_dm.loc[closest_after_idx, 'date']
                    selected_visit_num = after_dm.loc[closest_after_idx, 'visit_num']

        current_patient_best_values[measurement_val_col_prefix] = best_value
        current_patient_best_values[f'{measurement_val_col_prefix}_Date'] = best_date
        best_visit_num_for_measurement[measurement_val_col_prefix] = selected_visit_num # Store for later use


    # --- Special handling for Age_at_Exam ---
    # Get the visit number for the selected best BMI value
    bmi_visit_num = best_visit_num_for_measurement.get('BMI')
    if pd.notna(bmi_visit_num):
        # Construct the column name for Age_at_Exam for that specific visit
        age_at_exam_col_name = f'Age_at_Exam_visit_{int(bmi_visit_num)}'
        # Safely get the value for Age_at_Exam for the current patient's row
        age_at_exam_value = patient_data.get(age_at_exam_col_name)
        if pd.notna(age_at_exam_value):
            current_patient_best_values['Age_at_Exam'] = age_at_exam_value
        else:
            current_patient_best_values['Age_at_Exam'] = np.nan # If value is NaN, set to NaN
    else:
        current_patient_best_values['Age_at_Exam'] = np.nan # If no best BMI visit found, set Age_at_Exam to NaN


    best_visit_rows.append(current_patient_best_values)

    # Use a sequential counter for clearer progress reporting
    if (sequential_idx + 1) % 1000 == 0 or (sequential_idx + 1) == data_copy_filtered.shape[0]:
        print(f"Processed {sequential_idx + 1}/{data_copy_filtered.shape[0]} patients...")


# Create the final DataFrame
best_visit_df = pd.DataFrame(best_visit_rows)

print("\nProcessing complete.")
print("\nHead of the refined best_visit_df:")
print(best_visit_df.head())
print("\nShape of the refined best_visit_df:")
print(best_visit_df.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

def get_scores(y_pred, y_true):
    conf_matrix = confusion_matrix(y_true, y_pred)
    class_report = classification_report(y_true, y_pred, output_dict=True)
    accuracy = accuracy_score(y_true, y_pred)
    return conf_matrix, class_report, accuracy

def classification_report_str(class_report_dict):
    lines = []
    # Add header
    lines.append(f"{'':<10}{'precision':>9}{'recall':>9}{'f1-score':>9}{'support':>9}")
    lines.append(f"{'':<10}{'---------':>9}{'------':>9}{'--------':>9}{'-------':>9}")

    # Add class metrics
    for label, metrics in class_report_dict.items():
        if isinstance(metrics, dict):
            lines.append(f"{label:<10}{metrics['precision']:>9.2f}{metrics['recall']:>9.2f}{metrics['f1-score']:>9.2f}{metrics['support']:>9.0f}")

    # Add overall metrics
    for label in ['accuracy', 'macro avg', 'weighted avg']:
        if label in class_report_dict:
            metrics = class_report_dict[label]
            if label == 'accuracy': # Accuracy does not have precision, recall, f1-score
                lines.append(f"\n{label:<10}{'':>18}{metrics:>9.2f}{class_report_dict['macro avg']['support']:>9.0f}")
            else:
                lines.append(f"{label:<10}{metrics['precision']:>9.2f}{metrics['recall']:>9.2f}{metrics['f1-score']:>9.2f}{metrics['support']:>9.0f}")

    return "\n".join(lines)

def print_our_scores(conf_matrix, class_report_dict, accuracy):
    print("\nConfusion Matrix:")
    print(conf_matrix)
    print("\nClassification Report:")
    print(classification_report_str(class_report_dict)) # Convert dict to string for printing
    print(f"\nAccuracy: {accuracy:.4f}")

def plot_roc_curve(fpr, tpr, auc_score, title='ROC AUC Curve'):
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f'AUC = {auc_score:.3f}')
    plt.plot([0, 1], [0, 1], '--', color='grey')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc='lower right')
    plt.show()

def plot_confusion_matrix(conf_matrix):
    plt.figure(figsize=(6, 5))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()

print("\n--- Final Model Evaluation on the Held-Out Test Set (L1 Logistic Regression) ---")

# Make predictions on the held-out test set
y_pred_test_l1 = best_logreg_l1_model.predict(x_test)
y_prob_test_l1 = best_logreg_l1_model.predict_proba(x_test)[:, 1] # Probability of the positive class

# Evaluate and report metrics using the helper functions
test_conf_matrix_l1, test_class_report_dict_l1, test_accuracy_l1 = get_scores(y_pred_test_l1, y_test)
print_our_scores(test_conf_matrix_l1, test_class_report_dict_l1, test_accuracy_l1)

# Calculate ROC AUC score for the test set
test_roc_auc_l1 = roc_auc_score(y_test, y_prob_test_l1)
print(f"\nTest Set ROC AUC (L1 Logistic Regression): {test_roc_auc_l1:.4f}")

# Plot ROC AUC Curve for the test set
fpr_test_l1, tpr_test_l1, _ = roc_curve(y_test, y_prob_test_l1)
plot_roc_curve(fpr_test_l1, tpr_test_l1, test_roc_auc_l1, title='ROC AUC Curve for Tuned L1 Logistic Regression on Test Set')

# Plot Confusion Matrix for the test set
plot_confusion_matrix(test_conf_matrix_l1)
```

PyTorch architecture: training loop, stability fixes, tracking for learning curves

```python
class ClinicalNN(BaseEstimator, ClassifierMixin):
    _estimator_type = "classifier"

    def __init__(self, h1=64, h2=32, lr=0.001, dropout=0.3, epochs=100):
        self.h1, self.h2, self.lr = h1, h2, lr
        self.dropout, self.epochs = dropout, epochs
        self.model = None
        self.classes_ = np.array([0, 1])
        self.train_accuracy_list = []
        self.validation_accuracy_list = []

    def fit(self, X, y, X_val=None, y_val=None):
        # Ensure X and y are numeric arrays (handles output from preprocessor)
        X_t = t.as_tensor(np.array(X), dtype=t.float32)
        y_t = t.as_tensor(np.array(y), dtype=t.float32).reshape(-1, 1)

        self.model = nn.Sequential(
            nn.Linear(X_t.shape[1], self.h1), nn.ReLU(), nn.Dropout(self.dropout),
            nn.Linear(self.h1, self.h2), nn.ReLU(), nn.Dropout(self.dropout),
            nn.Linear(self.h2, 1), nn.Sigmoid()
        )

        optimizer = Adam(self.model.parameters(), lr=self.lr)
        loss_fn = nn.BCELoss()
        self.train_accuracy_list, self.validation_accuracy_list = [], []

        for epoch in range(self.epochs):
            self.model.train()
            optimizer.zero_grad()
            outputs = t.clamp(self.model(X_t), min=1e-7, max=1-1e-7)
            loss = loss_fn(outputs, y_t)
            loss.backward()
            t.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
            optimizer.step()

            # Tracking for Learning Curves
            self.model.eval()
            with t.no_grad():
                self.train_accuracy_list.append(((outputs >= 0.5).float() == y_t).float().mean().item())
                if X_val is not None and y_val is not None:
                    # X_val must be pre-processed
                    X_v_t = t.as_tensor(np.array(X_val), dtype=t.float32)
                    y_v_t = t.as_tensor(np.array(y_val), dtype=t.float32).reshape(-1, 1)
                    v_out = self.model(X_v_t)
                    self.validation_accuracy_list.append(((v_out >= 0.5).float() == y_v_t).float().mean().item())
        return self

    def predict_proba(self, X):
        self.model.eval()
        X_t = t.as_tensor(np.array(X), dtype=t.float32)
        with t.no_grad():
            prob_1 = self.model(X_t).numpy()
        return np.hstack([1 - prob_1, prob_1])

    def predict(self, X):
        return (self.predict_proba(X)[:, 1] >= 0.5).astype(int)
```

Pipline & GridSearch

```python
# Define the NN pipeline
nn_pipeline = ImbPipeline([
    ('preprocessor', preprocessing_transformer),
    ('undersampler', RandomUnderSampler(random_state=42)),
    ('nn', ClinicalNN())
])

param_grid = {
    'nn__h1': [32, 64],
    'nn__h2': [16, 32],
    'nn__lr': [0.01, 0.001, 0.0001],
    'nn__dropout': [0.2, 0.3, 0.5],
    'nn__epochs': [100]
}

n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

print(f"\nPerforming {n_splits}-fold Stratified Cross-Validation with GridSearchCV for Neural Network...")

grid_search_nn = GridSearchCV(
    estimator=nn_pipeline,
    param_grid=param_grid_nn,
    cv=skf,
    scoring='roc_auc',
    n_jobs=1,
    verbose=1
)

grid_search_nn.fit(x_train_nn, y_train_nn)

print("\nGridSearchCV for Neural Network completed.")
```
